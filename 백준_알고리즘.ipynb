```python
# 시간 초과, 원하는 답은 나옴
n = int(input())

a = list(map(int, input().split()))

b, c = map(int, input().split())

ans = 0

for i in range (n):
  a[i] -= b
  ans += 1

while True:
  check = 0
  for i in range (n):
    if a[i] > 0:
      a[i] -= c
      ans += 1
      check += 1
  if check == 0:
    break

print(ans)
```

```python
# 올림을 이용하기 위해 math 호출
import math

n = int(input())

a = list(map(int, input().split()))

b, c = map(int, input().split())

ans = 0

# 총감독관 배치
for i in range (n):
  a[i] -= b
  ans += 1

# 부감독관 배치 / 총 명수에 (남은 인원 / 부감독 커버 인원)값을 올림 한것을 추가한다.
for i in range (n):
  if a[i] > 0:
    ans += math.ceil(a[i] / c)
    a[i] = 0


print(ans)
```
