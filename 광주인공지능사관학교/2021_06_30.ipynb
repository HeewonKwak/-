# **카카오 코드 페스티벌 2018 예선 A번 - 상금 헌터**

https://www.acmicpc.net/problem/15953

```python
n = int(input())

rewards = []

for i in range(n):
  reward = 0
  a, b = map(int, input().split())
  if a == 0:
    reward += 0
  elif a == 1:
    reward += 500
  elif 4 > a > 1:
    reward += 300
  elif 7 > a > 3:
    reward += 200
  elif 11 > a > 6:
    reward += 50
  elif 16 > a > 10:
    reward += 30
  elif 22 > a > 15:
    reward += 10
  else:
    reward += 0
  if b == 0:
    reward += 0
  elif b == 1:
    reward += 512
  elif 4 > b > 1:
    reward += 256
  elif 8 > b > 3:
    reward += 128
  elif 16 > b > 7:
    reward += 64
  elif 32 > b > 15:
    reward += 32
  else:
    reward += 0
  rewards.append(reward)

for i in range(n):
  print(rewards[i]*10000)
```

```python
a, b, r = map(int, input().split())
n = int(input())

checks = []

for i in range (n):
  check = 0
  x, y = map(int, input().split())
  region = (x - a)**2 + (y - b)**2
  if region >= r**2:
    check = 1
  else:
    check = 0
  checks.append(check)

for j in checks:
  if j == 1:
    print("silent")
  else:
    print("noisy")
```

```python
# 강사님 코드
def check_sound_status(a,b,r,x,y):
  if (x-a)**2 + (y-b)**2 >= r**2:
    return "silent"
  return "noisy"

a, b, r = map(int, input().split())
n = int(input())

for _ in range(n):
  x, y = map(int, input().split())
  print(check_sound_status(a,b,r,x,y))
```
