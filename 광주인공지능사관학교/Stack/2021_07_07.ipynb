```python
n = int(input())

ans_list = []
list_len = 0
for i in range(n):
  ment = input()
  if ment == "top":
    if list_len == 0:
      print(-1)
    else:
      print(ans_list[list_len-1])
  elif ment == "size":
    print(list_len)
  elif ment == 'pop':
    if list_len == 0:
      print(-1)
    else:
      list_len -= 1
      print(ans_list[list_len])
      ans_list.pop()
  elif ment == 'empty':
    if list_len == 0:
      print(1)
    else:
      print(0)
  else:
    list_len += 1
    command, num =  ment.split()
    num = int(num)
    ans_list.append(num)
```

```python
n = int(input())

ans_list = []
for i in range(n):
  ment = input()
  if ment == "top":
    if len(ans_list) == 0:
      print(-1)
    else:
      print(ans_list[-1])
  elif ment == "size":
    print(len(ans_list))
  elif ment == 'pop':
    if len(ans_list) == 0:
      print(-1)
    else:
      print(ans_list[-1])
      ans_list.pop()
  elif ment == 'empty':
    if len(ans_list) == 0:
      print(1)
    else:
      print(0)
  else:
    command, num =  ment.split()
    num = int(num)
    ans_list.append(num)
```
